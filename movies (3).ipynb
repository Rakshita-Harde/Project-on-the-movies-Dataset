```python
# Number of unique genres in the data set

genres_col = df['genres']
# Empty set to store unique genres
unique_genres = set()
# Iterate over each row in the genres column
for genres_list in genres_col:
    genres = eval(genres_list)  # Convert the string representation of list to a list
    unique_genres.update(genres)  # Add the genres to the set
    
len(unique_genres)
```

```python
# Top 5 moies in every genre

for genre in unique_genres:
    movies = df[df['genres'].str.contains(genre)].sort_values('popularity', ascending=False).head(5)
    movie_title = movies['title']
    print(genre)
    for i ,title in enumerate(movie_title):
        print(i+1, title)
    print("\n")
```

```python
#Top 5 production companies with maximum movie count

companies_column = df['production_companies']

# Create an empty dictionary to store company names and their movie counts
company_counts = {}

# Iterate over each row in the companies column
for companies_list in companies_column:
    companies = eval(companies_list)  # Convert the string representation of list to a list
    for company in companies:
        if company in company_counts:
            company_counts[company] += 1  # Increment the movie count
        else:
            company_counts[company] = 1  # Add the company with initial movie count

sorted_companies = sorted(company_counts.items(), key=lambda x: x[1], reverse=True)

top_5_companies = sorted_companies[:5]

for company, count in top_5_companies:
    print(company, ": ", count)
```

```python
#Most popular movie in every 5 years with their popularity

# Convert the release_date column to datetime
df['release_date'] = pd.to_datetime(df['release_date'])

# Create a new column for the release half_decade
df['release_hd'] = (df['release_date'].dt.year // 5) * 5

popular_movies = df.groupby('release_hd').apply(lambda x: x.nlargest(1, 'popularity')) # 1 is for no. of rows we want in every decade

for i, r in popular_movies.iterrows():
    print(f"Year: {r['release_hd']}-{r['release_hd']+4}")
    print("Movie:", r['title'])
    print("Popularity:", r['popularity'])
    print("\n")
```

# Strong Positive Correlations
1. Revenue vs Profit (0.98): Higher revenue almost guarantees higher profit.  
  *Recommendation:* Focus on expanding revenue sources like which increases the profit

2. Revenue vs Vote Count (0.75): Popular movies (in terms of votes) tend to earn more.  
  *Recommendation:* Increase advertisement and public engagement before and after release.

3. Budget vs Revenue (0.74): Bigger budgets are often associated with higher earnings.  
  *Recommendation:* As the bigger budget movie tends to earn more , so producer can plan the budget accordingly

4. Vote Count vs Profit (0.72): More audience engagement leads to better profit.  
  *Recommendation:* Boost user interaction on social platforms.

5. Budget vs Vote Count (0.60): High-budget movies get more votes, likely due to wider reach.
   *Recommendation:* social reach can improvevote count.

# Moderate Correlations
1. Vote Average has weak correlation with financial success.
2. Popularity is weakly correlated with most features.

# Negative Correlation
1. Budget vs Vote Average (-0.08)
Movies with higher budgets do not necessarily receive higher average ratings. This may indicate that spending more doesn't always guarantee better audience reception.

2. Vote Average vs Popularity (-0.036)
Well-rated movies are not always popular. Critically appreciated films might still be less known to the general audience.

3. Vote Average vs Profit, Revenue, Vote Count
Vote average shows weak or no correlation with profit, revenue, and vote count.
